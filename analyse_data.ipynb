```python
import pandas as pd
```

```python
header_lines = []
header_info = {}
csv_file = "raw_data/arbin5/1721734451_17_1_20240723_BMW_IT0_T25.csv"

with open(csv_file, 'r') as f:
    for i, line in enumerate(f):
        stripped = line.strip()
        if "Data Field:" in line:
            skiprows = i + 1
            break
        header_lines.append(stripped)

        if ':' in stripped:
            key, value = stripped.split(":", 1)
            header_info[key.strip()] = value.strip()
```

```python
data = pd.read_csv(csv_file, skiprows=skiprows, sep='\t')
data
```

```python
data = data.drop(0)
data.reset_index(drop=True, inplace=True)
data
```

```python
data['Current'] = data['Current'].astype(float)
data
```

```python
min_current = data['Current'].min()
max_current = data['Current'].max()

print(f"Minimum Current: {min_current}")
print(f"Maximum Current: {max_current}")
```
